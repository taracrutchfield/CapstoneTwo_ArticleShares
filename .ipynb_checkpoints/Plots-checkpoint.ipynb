```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

c1 = '#6DB1BF' # cyan
c2 = '#315659' # blue
c3 = '#F98948' # orange
c4 = '#C0DA74' # green

path = '../../../Desktop/Plots/'

plt.rcParams['figure.figsize'] = 11, 5
plt.rcParams['figure.dpi'] = 200
plt.rcParams.update({'font.size': 12})
```

```python
data = pd.read_csv('OnlineNewsPopularity.csv').set_index('url')

fig = plt.figure()

ax1 = plt.subplot(122)
median = 'Shares Median: %d' % data.shares.median()
plt.hist(data.shares,range=(0,np.percentile(data.shares,[0,90])[1]),bins=10,color=c1,rwidth=.95)
plt.axvline(data.shares.median(),c=c3,linestyle='--',label=median)
plt.legend();

ax2 = plt.subplot(121)
plt.hist(data.shares,bins=10,color=c1,rwidth=.95);

fig.text(0.5, 0.03, 'Article Shares', ha='center', va='center')
fig.text(0.05, 0.5, 'Num Articles', ha='center', va='center', rotation='vertical')

ax1.set_title('Shares Distribution - 75%')
ax2.set_title('Shares Distribution')

fig.savefig(path+'ArticleHist.png', dpi=200)
```
